```python
import datetime

start = datetime.datetime.strptime("21-06-2001", "%d-%m-%Y")
end = datetime.datetime.strptime("31-12-3000", "%d-%m-%Y")
date_generated = [start + datetime.timedelta(days=x) for x in range(0, (end-start).days)]

for date in date_generated:
    dates = date.strftime("%-m-%-d-%Y")
    dates = dates.replace('-','')
    if len(dates)==6:
        if dates[0]==dates[-1] and dates[1]==dates[-2] and dates[2]==dates[-3]:
            dates= date.strftime("%-m-%-d-%Y")
            print(dates)
    elif len(dates)==7:
        if dates[0]==dates[-1] and dates[1]==dates[-2] and dates[2]==dates[-3]:
            dates= date.strftime("%-m-%-d-%Y")
            print(dates)

    else: 
        if dates[0]==dates[-1] and dates[1]==dates[-2] and dates[2]==dates[-3] and dates[3]==dates[-4] :
            dates= date.strftime("%-m-%-d-%Y")
            print(dates)
```
